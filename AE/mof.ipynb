```python
#lab-1
import numpy as np 
import matplotlib.pyplot as plt 

theta = np.linspace(0,np.pi,360)

R = np.abs(np.sin(theta))

ax = plt.subplot(projection='polar')
ax.plot(theta,R)
ax.set_title("2D Dipole Antenna")
plt.show()

theta = np.linspace(0,np.pi,1000)
phi = np.linspace(0,2*np.pi,1000)
theta,phi = np.meshgrid(theta,phi)

R = np.abs(np.sin(theta))

X = R * np.sin(theta) * np.cos(phi)
Y = R * np.sin(theta) * np.sin(phi)
Z = R * np.cos(theta)

ax = plt.subplot(projection='3d')
ax.plot_surface(X,Y,Z)
ax.set_title("3D Dipole Antenna")
plt.show()
```

```python
#lab-2
import numpy as np 
import matplotlib.pyplot as plt 

theta = np.linspace(0,2*np.pi,360)
R = np.ones_like(theta)

ax = plt.subplot(projection='polar')
ax.plot(theta,R)
ax.set_title("2D Isotropic Antenna")
plt.show()

theta = np.linspace(0,np.pi,1000)
phi = np.linspace(0,2*np.pi,1000)
theta,phi = np.meshgrid(theta,phi)

R = np.ones_like(theta)

X = R * np.sin(theta) * np.cos(phi)
Y = R * np.sin(theta) * np.sin(phi)
Z = R * np.cos(theta)

ax = plt.subplot(projection='3d')
ax.plot_surface(X,Y,Z,cmap='plasma')
ax.set_title("3D Isotropic Antenna")
plt.show()
```

```python
#lab-3
import numpy as np 
import matplotlib.pyplot as plt 

n = 10

theta = np.linspace(-np.pi/2,np.pi/2,360)

R = (np.abs(np.cos(theta))) ** n

ax = plt.subplot(projection='polar')
ax.plot(theta,R)
ax.set_title("2D Horn Antenna")
plt.show()

theta = np.linspace(0,np.pi,1000)
phi = np.linspace(0,2*np.pi,1000)
theta , phi = np.meshgrid(theta,phi)

#R = np.abs(np.cos(theta))
R = np.cos(theta) ** n
R[theta > np.pi / 2] = 0
R = np.abs(R)

X = R * np.sin(theta) * np.cos(phi)
Y = R * np.sin(theta) * np.sin(phi)
Z = R * np.cos(theta)

ax = plt.subplot(projection='3d')
ax.plot_surface(X,Y,Z,cmap='inferno')
ax.set_title("3D Horn Antenna")
plt.show()
```

```python
#lab-4
import numpy as np 
import matplotlib.pyplot as plt 

N = 10 

lam = 1
k = 2*np.pi / lam
d = lam / 2 

theta = np.linspace(0,np.pi,360)

num = np.sin(N*k*d*np.cos(theta)/2)
den = N * np.sin(k*d* np.cos(theta)/2)

AF = np.where(den == 0,1,num/den)

AF = np.abs(AF)

ax = plt.subplot(projection='polar')
ax.plot(theta,AF)
ax.set_title("2D Linear Array Antenna")
plt.show()

theta = np.linspace(0,np.pi,1000)
phi = np.linspace(0,2*np.pi,1000)
theta,phi = np.meshgrid(theta,phi)

num = np.sin(N * k* d * np.cos(theta)/2)
den = N * np.sin(k*d*np.cos(theta)/2)

AF = np.where(den == 0,1,num/den)
AF = np.abs(AF)

X = AF * np.sin(theta) * np.cos(phi)
Y = AF * np.sin(theta) * np.sin(phi)
Z = AF * np.cos(theta)

ax = plt.subplot(projection='3d')
ax.plot_surface(X,Y,Z,cmap='viridis')
ax.set_title("3D Linear Array Antenna")
plt.show()
```

```python
#lab-5
import numpy as np 
import matplotlib.pyplot as plt 

N = 3   # just change here for 5 , 7 elements
lam = 1
d = lam / 2
k = 2*np.pi / lam 

theta = np.linspace(0,np.pi,360)

AF = np.zeros_like(theta,dtype=complex)

for n in range(N):
    AF += np.exp(1j * n * k * d * np.cos(theta))

AF_norm = np.abs(AF) / np.max(np.abs(AF))

ax = plt.subplot(projection='polar')
ax.plot(theta,AF_norm)
ax.set_title("2D Broadside Arrar Antenna -3 elements")
plt.show()

theta = np.linspace(0,np.pi,1000)
phi = np.linspace(0,2*np.pi,1000)
theta,phi = np.meshgrid(theta,phi)

AF = np.zeros_like(theta,dtype=complex)

for n in range(N):
    AF += np.exp(1j * n * k * d * np.cos(theta))

AF_norm = np.abs(AF) / np.max(np.abs(AF))

X = AF_norm * np.sin(theta) * np.cos(phi)
Y = AF_norm * np.sin(theta) * np.sin(phi)
Z = AF_norm * np.cos(theta)

ax = plt.subplot(projection='3d')
ax.plot_surface(X,Y,Z,cmap='inferno')
ax.set_title("3D Broadside Array Antenna -3 elements")
plt.show()
```

```python
#lab-6
import numpy as np 
import matplotlib.pyplot as plt 

theta = np.linspace(0,2*np.pi,360)

lam = 1
d = lam / 2
k = 2*np.pi / lam 

AF = 2 * np.cos((k * d / 2) * np.cos(theta))
# AF = 2 * np.sin((k*d/2)*np.cos(theta)) for equal amplitude and opposite phase

AF_norm = np.abs(AF) / np.max(np.abs(AF))

ax = plt.subplot(projection='polar')
ax.plot(theta,AF_norm)
ax.set_title("2D Far-Field Pattern\nEqual Amplitude and Phase")
plt.show()

theta = np.linspace(0,np.pi,1000)
phi = np.linspace(0,2*np.pi,1000)
theta,phi = np.meshgrid(theta,phi)

AF = 2 * np.cos((k * d / 2) * np.cos(theta))
# AF = 2 * np.sin((k*d/2)*np.cos(theta)) for equal amplitude and opposite phase

AF_norm = np.abs(AF) /np.max(np.abs(AF))

X = AF_norm * np.sin(theta) * np.cos(phi)
Y = AF_norm * np.sin(theta) * np.sin(phi)
Z = AF_norm * np.cos(theta)

ax = plt.subplot(projection='3d')
ax.plot_surface(X,Y,Z,cmap='viridis')
ax.set_title("3D Far-Field Pattern\nEqual Amplitude and Phase")
plt.show()
```

```python
#lab-7
import numpy as np 
import matplotlib.pyplot as plt 

theta = np.linspace(0,2*np.pi,360)

lam = 1
d = lam / 2
k = 2*np.pi / lam

A1 = 1.0 
A2 = 0.6
psi = (k * d /2) * np.cos(theta)

AF = A1*np.exp(1j * psi) - A2*np.exp(-1j * psi)

AF_norm = np.abs(AF) / np.max(np.abs(AF))

ax = plt.subplot(projection='polar')
ax.plot(theta,AF_norm)
ax.set_title("2D Far-Field Pattern\nUnequal Amplitude and Opposite Phase")
plt.show()

theta = np.linspace(0,np.pi,1000)
phi = np.linspace(0,2*np.pi,1000)
theta,phi = np.meshgrid(theta,phi)

psi = (k * d / 2) * np.cos(theta)

AF = A1 * np.exp(1j * psi) - A2 * np.exp(-1j * psi)

AF_norm = np.abs(AF) / np.max(np.abs(AF))

X = AF_norm * np.sin(theta) * np.cos(phi)
Y = AF_norm * np.sin(theta) * np.sin(phi)
Z = AF_norm * np.cos(theta)

ax = plt.subplot(projection='3d')
ax.plot_surface(X,Y,Z,cmap='inferno')
ax.set_title("3D Far-Field Pattern\nUnequal Amplitude and Opposite Phase")
plt.show()
```
